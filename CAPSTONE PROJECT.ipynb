## AUTOMOBILE DATA – PREPROCESSING & VISUALIZATION PROJECT

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## 2.LOAD DATASET

```python
df = pd.read_csv(r"D:\ITVEDANT\applied python\Automobile_data (2).csv")
print(df.head())
```

```python
# Replace missing values
df.replace('?', pd.NA, inplace=True)

# Convert columns to numeric
df['normalized-losses'] = pd.to_numeric(df['normalized-losses'], errors='coerce')
df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')
df['price'] = pd.to_numeric(df['price'], errors='coerce')

# Drop missing values
df = df.dropna()
```

## 2. Correlation Heatmap

```python
corr = df[['price','engine-size','horsepower','width','height','city-mpg']].corr()

sns.heatmap(corr, annot=True)

plt.title("Correlation Heatmap")
plt.show()
plt.figure(figsize=(8,5))
```

## 3. Average Price by Brand

```python
brand_price = df.groupby('make')['price'].mean().sort_values()

brand_price.plot(kind='bar')

plt.title("Average Price by Brand")
plt.xlabel("Brand")
plt.ylabel("Price")

plt.show()
plt.figure(figsize=(8,5))
```

## 👉 Insight:

## Luxury brands → higher price

## 4. Boxplot (Body Style vs Price)

```python
sns.boxplot(x='body-style', y='price', data=df)

plt.title("Price by Body Style")
plt.show()
plt.figure(figsize=(8,5))
```

## 5. Horsepower vs Price

```python
sns.scatterplot(x='horsepower', y='price', data=df)

plt.title("Horsepower vs Price")
plt.show()
plt.figure(figsize=(8,5))
```

## 6. Engine Size vs Mileage

```python
sns.scatterplot(x='engine-size', y='city-mpg', data=df)

plt.title("Engine Size vs Mileage")
plt.show()
plt.figure(figsize=(8,5))
```

## 7. Fuel Type vs Price

```python
sns.barplot(x='fuel-type', y='price', data=df)

plt.title("Fuel Type vs Price")
plt.show()
plt.figure(figsize=(8,5))
```

## 8. Drive Wheels vs Price

```python
sns.boxplot(x='drive-wheels', y='price', data=df)

plt.title("Drive Wheels vs Price")
plt.show()
plt.figure(figsize=(8,5))
```

## 9. Pairplot (Important for Marks)

```python
sns.pairplot(df[['price','horsepower','engine-size','city-mpg']])
plt.show()
plt.figure(figsize=(8,5))
```

## 1. Correlation Heatmap – Price vs Numerical Features

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

corr = df.select_dtypes(include='number').corr()

sns.heatmap(corr, annot=True)

plt.title("Correlation Heatmap")
plt.show()
plt.figure(figsize=(8,5))
```

### Insight:
### Price is highly correlated with engine size and horsepower
### Price is negatively related to fuel efficiency (MPG)
